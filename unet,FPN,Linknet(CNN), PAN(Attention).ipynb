```python
def lee_sigma_filter(image, window_size=7, sigma_threshold=0.8):
    image = image.astype(np.float32)
    rows, cols = image.shape
    denoised_image = np.zeros_like(image)
    half_window = window_size // 2
    for i in range(half_window, rows - half_window):
        for j in range(half_window, cols - half_window):
            window = image[i - half_window:i + half_window + 1, j - half_window:j + half_window + 1]
            window_mean = np.mean(window)
            window_variance = np.var(window)
            if window_variance < sigma_threshold:
                denoised_image[i, j] = window_mean
            else:
                denoised_image[i, j] = image[i, j]
    return denoised_image
```

```python
denoised_s1_images = []
for image in s1_images:
    vv_band = image[:, :, 0] 
    vh_band = image[:, :, 1]  
    denoised_vv = lee_sigma_filter(vv_band)
    denoised_vh = lee_sigma_filter(vh_band)
    denoised_s1_images.append(np.stack([denoised_vv, denoised_vh], axis=-1))
denoised_s1_images = np.array(denoised_s1_images)
```

```python

from sklearn.metrics import cohen_kappa_score, confusion_matrix, classification_report
import numpy as np
def calculate_metrics(y_true, y_pred, num_classes):
    if isinstance(y_true, tf.Tensor):
        y_true = y_true.numpy()
    if isinstance(y_pred, tf.Tensor):
        y_pred = y_pred.numpy()
    if len(y_true.shape) > 1 and y_true.shape[-1] == num_classes:
        y_true = np.argmax(y_true, axis=-1)
    if len(y_pred.shape) > 1 and y_pred.shape[-1] == num_classes:
        y_pred = np.argmax(y_pred, axis=-1)
    y_true_flattened = y_true.flatten()
    y_pred_flattened = y_pred.flatten()
    kappa = cohen_kappa_score(y_true_flattened, y_pred_flattened)
    oa = np.sum(y_pred_flattened == y_true_flattened) / float(y_true_flattened.size)
    conf_matrix = confusion_matrix(y_true_flattened, y_pred_flattened, labels=np.arange(num_classes))
    row_sums = np.sum(conf_matrix, axis=1) 
    correct_predictions = np.diag(conf_matrix) 
    valid_classes = row_sums > 0  
    if np.any(valid_classes):
        aa = np.mean(correct_predictions[valid_classes] / row_sums[valid_classes])
    else:
        aa = np.nan  
    class_report = classification_report(
        y_true_flattened, y_pred_flattened, labels=np.arange(num_classes), zero_division=1
    )
    return {
        "Cohens Kappa": kappa,
        "Overall Accuracy (OA)": oa,
        "Average Accuracy (AA)": aa,
        "Classification Report": class_report
    }
```

```python
from keras import layers
class CustomMaskLayer(layers.Layer):
    def __init__(self, **kwargs):
        super(CustomMaskLayer, self).__init__(**kwargs)
    def build(self, input_shape):
        pass
    def call(self, inputs):
        return inputs
    def get_config(self):
        config = super(CustomMaskLayer, self).get_config()
        return config
import tensorflow as tf
from keras import backend as K

def focal_loss_plus_jaccard_loss(y_true, y_pred):
    alpha = 0.25
    gamma = 2.0
    focal_loss = -alpha * K.pow(1 - y_pred, gamma) * y_true * K.log(y_pred)
    intersection = K.sum(y_true * y_pred, axis=(1, 2, 3))
    union = K.sum(y_true, axis=(1, 2, 3)) + K.sum(y_pred, axis=(1, 2, 3)) - intersection
    jaccard_loss = 1 - (intersection + 1) / (union + 1)
    return K.mean(focal_loss + jaccard_loss)
```

```python
import torch
from torch import optim
from torch.optim.lr_scheduler import ReduceLROnPlateau
from torch.utils.data import DataLoader, TensorDataset
import segmentation_models_pytorch as smp
BACKBONE = 'resnet34'
ENCODER_WEIGHTS = 'imagenet'
INPUT_CHANNELS = 2
NUM_CLASSES = 5
model_pan_500_5 = smp.PAN(
    encoder_name=BACKBONE, 
    encoder_weights=ENCODER_WEIGHTS, 
    in_channels=INPUT_CHANNELS, 
    classes=NUM_CLASSES, 
    activation='softmax'
)
optimizer = optim.Adam(model_pan_500_5.parameters(), lr=0.001)
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=5, min_lr=1e-6)
def combined_focal_jaccard_loss(y_true, y_pred):
    focal_loss = smp.losses.FocalLoss(mode='multiclass')
    jaccard_loss = smp.losses.JaccardLoss(mode='multiclass')
    return focal_loss(y_pred, y_true) + jaccard_loss(y_pred, y_true)
y1_val5=y1_val5.numpy()
y1_val5 = torch.argmax(torch.tensor(y1_val5), dim=-1)  
y1_train5 = torch.argmax(torch.tensor(y1_train5), dim=-1) 
train_dataset = TensorDataset(torch.tensor(X1_train5), y1_train5)
val_dataset = TensorDataset(torch.tensor(X1_val5), y1_val5)
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=64)
import psutil
import time
process = psutil.Process(os.getpid())
before_training_memory = process.memory_info().rss
start_time = time.time()
# Training Loop
num_epochs = 100
for epoch in range(num_epochs):
    model_pan_500_5.train()
    train_loss = 0
    for X_batch, y_batch in train_loader:
        optimizer.zero_grad()
        outputs = model_pan_500_5(X_batch)
        loss = combined_focal_jaccard_loss(y_batch, outputs)
        loss.backward()
        optimizer.step()
        train_loss += loss.item()

    model_pan_500_5.eval()
    val_loss = 0
    with torch.no_grad():
        for X_batch, y_batch in val_loader:
            outputs = model_pan_500_5(X_batch)
            loss = combined_focal_jaccard_loss(y_batch, outputs)
            val_loss += loss.item()

    # Scheduler step
    scheduler.step(val_loss)

    print(f"Epoch [{epoch+1}/{num_epochs}], Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}")
end_time = time.time()
after_training_memory = process.memory_info().rss
```
